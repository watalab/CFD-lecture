```julia
using PyPlot
```

```julia
# parameter
c = 1.0
dt = 0.05
dx = 0.1
jmax = 21
nmax = 12
```

# central differencing

```julia
x = zeros(jmax)
q = zeros(jmax)
qold = zeros(jmax)

fig = plt.figure(figsize = (5, 4))
ax1 = fig.add_subplot(111)

for j=1:jmax
    x[j] = dx*(j-1)
    if j < jmax/2
        q[j] = 1.0
    end
end

ax1.plot(x, q, marker="o", label="time step="*string(0))

for n = 1:nmax
    for j = 1:jmax
        qold[j] = q[j]
    end
    
    for j = 2:jmax-1
        q[j] = qold[j] - dt*c*(qold[j+1] - qold[j-1])/(2dx)
    end
    if n%3 == 0
        ax1.plot(x, q, marker="o", label="time step="*string(n))
    end
end
        

ax1.set_xlabel("x")
ax1.set_ylabel("q")
ax1.grid(linestyle="dashed", linewidth=0.5)
ax1.legend()
#plt.show()
    
```

# upwind

```julia
x = zeros(jmax)
q = zeros(jmax)
qold = zeros(jmax)

fig = plt.figure(figsize = (5, 4))
ax2 = fig.add_subplot(111)

for j=1:jmax
    x[j] = dx*(j-1)
    if j < jmax/2
        q[j] = 1.0
    end
end

ax2.plot(x, q, marker="o", label="time step="*string(0))

for n = 1:nmax
    for j = 1:jmax
        qold[j] = q[j]
    end
    
    for j = 2:jmax-1
        q[j] = qold[j] - dt*c*(qold[j] - qold[j-1])/dx
    end
    if n%3 == 0
        ax2.plot(x, q, marker="o", label="time step="*string(n))
    end
end
        

ax2.set_xlabel("x")
ax2.set_ylabel("q")
ax2.grid(linestyle="dashed", linewidth=0.5)
ax2.legend()
#plt.show()
    
```
